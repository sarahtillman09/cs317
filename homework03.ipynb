## Part 3: Autograd

```python
import autograd.numpy as np
from autograd import grad
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def banana(z):
    return np.array(100 * (z[1] - z[0] ** 2) ** 2 + (1 - z[0]) ** 2)
```

### Problem 3.1 (Exam Style)

```python
from autograd import jacobian
banana_gradient = grad(banana)
banana_hessian = jacobian(banana_gradient)
```

```python
z0 = np.linspace(-2., 2., 101)

g = np.empty(shape=(len(z0),2),dtype='object')
H = np.empty(shape=(len(z0),3),dtype='object')
f = np.empty(shape=(len(z0)),dtype='float')

for z in range(len(z0)):
    f[z] = banana(np.array([z0[z], 1.]))
    g[z] = banana_gradient(np.array([z0[z], 1.]))
    H[z] = np.compress([True, True, False, True], 
                       banana_hessian(np.array([z0[z],1.])))

def plotting(vals, k, title, legend=None):
    plt.subplot(1,3, k)
    plt.plot(z0, vals)
    if legend != None:
        plt.legend(legend)
    plt.title(title)


plt.figure(figsize=(18, 5), constrained_layout=True)
plotting(f, 1, "Banana")
plotting(g, 2, "Gradient of Banana", ["g0", "g1"])
plotting(H, 3, "Hessian of Banana", ["H00", "H01", "H11"])
plt.show()
```

## Part 4: Gradient Descent

```python
from scipy import optimize
from numpy import linalg as npl


def line_search(f, g, z0, f0, g0, state=None):
    outcome = optimize.line_search(f, g, z0, -g0, g0, f0)
    alpha, f1 = outcome[0], outcome[3]
    evaluations = outcome[1]
    if alpha is None:
        alpha, f1 = 0., f0
    z1 = z0 - g0 * alpha
    return z1, f1, evaluations, state
```

```python
def gd(f, g, z0, step_function=line_search, state=None, max_evaluations=1000,
       min_step=1.e-8, min_gradient=1.e-6):
    evaluations, h = 0, []
    while True:
        f0, g0 = f(z0), g(z0)
        if not len(h):
            h.append((z0, f0))
        evaluations += 1
        if npl.norm(g0) < min_gradient:
            z1, f1 = z0, f0
            break
        z1, f1, n_eval, state = step_function(f, g, z0, f0, g0, state=state)
        evaluations += n_eval
        h.append((z1, f1))
        if npl.norm(z1 - z0) < min_step or evaluations > max_evaluations:
            break
        z0 = z1
    return z1, f1, evaluations, h
```

```python
def momentum(f, g, z0, f0, g0, state=None):
    if state is None:
        state = {'alpha': 0.001, 'v0': 0., 'mu': 0.9}
    v1 = state['mu'] * state['v0'] - g0 * state['alpha']
    z1 = z0 + v1
    state['v0'] = v1
    return z1, f(z1), 1, state
```

### Problem 4.1

```python
def fixed(f, g, z0, f0, g0, state=None):
    if state is None:
        state = {'alpha':0.001, 'v0':0., 'mu': 0.}
    if len(state) == 1:
        a = state['alpha']
        state = {'alpha':a, 'v0':0., 'mu':0.}
    return momentum(f, g, z0, f0, g0, state)
```

```python
z_0 = np.array((-1.2, 1.))
z_star = np.array([1., 1.])
```

```python
steps = ((line_search, None), 
         (fixed, {'alpha': 0.001}), (fixed, {'alpha': 0.005}),
          (momentum, {'alpha': 0.001, 'v0': 0., 'mu': 0.2}),
          (momentum, {'alpha': 0.001, 'v0': 0., 'mu': 0.5}),
          (momentum, {'alpha': 0.001, 'v0': 0., 'mu': 0.93}))
```

```python
from numpy import array2string


def plot_contours(f, z_ast, rect):
    assert len(z_ast) == 2, 'can only plot in two dimensions'
    n = 101
    xs = np.linspace(rect[0], rect[1], n)
    ys = np.linspace(rect[2], rect[3], n)
    fs = np.array([[f(np.array([x, y])) for x in xs] for y in ys])
    plt.contour(xs, ys, fs, 20, colors='grey', linewidths=1)
    plt.plot(z_ast[0], z_ast[1], 'ro')
    plt.axis('scaled')


def fa(a, p=2):
    return array2string(np.array(a), precision=p, separator=', ')
```

```python
def plot_history_path(h, f, z_ast, n_evals, state, rect=(-2., 2., -2., 2.)):
    assert len(z_ast) == 2, 'can only plot in two dimensions'
    plot_contours(f, z_ast, rect)
    plt.plot([p[0][0] for p in h], [p[0][1] for p in h])
    plt.plot(h[0][0][0], h[0][0][1], 'bo')
    z_last, f_last = h[-1][0], h[-1][1]
    ft = 'f({}) = {:.2g}; {} steps; {} evals'
    title = ft.format(fa(z_last), f_last, len(h), n_evals)
    if state is not None:
        s = state.copy()
        try:
            del s['v0']
        except KeyError:
            pass
        st = '; '.join(['{}={:.3g}'.format(name, value) for name, value in s.items()])
        title = '\n'.join((title, st))
    plt.title(title)
```

```python
plt.figure(figsize=(18, 11))
for i in range(0,6):
    plt.subplot(2, 3, i + 1)
    h = gd(banana, banana_gradient, 
           z_0, step_function = steps[i][0], 
           state = steps[i][1])
    plot_history_path(h[3], banana, z_star, h[2], state=steps[i][1])
plt.show()
```

### Problem 4.2 (Exam Style)

Out of all 3 methods, the momentum method appears to give the closest approximation to $\mathbf{z^*}$ since with $\mu = 0.93$, the final value for $f(\mathbf{\hat z})$ is less than 0.0001 which is much smaller than the final $f(\mathbf{\hat z})$ for the line search method ($f(\mathbf{\hat z})$ at 0.022) and the final $f(\mathbf{\hat z})$ for the fixed step method at either $\alpha$. However, in practice it would likely be more difficult to pick a value for $\mu$ since we don't know the actual value of the minimum we're trying to approximate. Picking too large a value for $\mu$ can result in the $\alpha$ for steps neat the minimum being too large and missing the minimum. At the same time, having a small value for $\mu$ results in the approximation of the minimum being less accurate.

The fixed step method is computationally and conceptually the most straightforward, but provided a less accurate approximation for the minimum than the other 2 methods. Increasing the value for $\alpha$ did not make the approximation more accurate and actually resulted in the step function reaching a shallow minimum before going in the opposite direction as this method did not allow for shallow minima to be passed over like in the momentum method.

The line search method provided a better approximation than the fixed step method, but not quite as accurate as the momentum method with $\mu = 0.93$. It's also important to note that this method used fewer steps at 125 steps compared to the 502 steps used in all iterations for the fixed step and momentum methods. The line search method also did not require specification of parameters like the other 2 methods did which could be useful when not much is known about the underlying function.
